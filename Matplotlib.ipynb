```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# see plots inline within notebook. outside notebooks you'd have to write > plt.show()
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# there are two ways to plot.
# 1: Functional method
# 2: Object oriented method
```

```python
# Functional method:
plt.plot(x, y)
```

```python
# adding labels
plt.xlabel('x')
plt.ylabel('y')
plt.title('simple x y plot')
plt.plot(x, y, 'r')
```

```python
# multiple plots (as a grid) at once using subplot(nrows, ncols, plot_number)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# Object oriented method using a figure object. # A figure object can be imagined as a blank canvas.
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.set_title('simple object oriented plot')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.plot(x, y)
```

```python
# Nested plots using object oriented method
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([.1, .5, .4, .4])

axes1.set_title('large plot')
axes1.plot(x, y)

axes2.set_title('inner plot')
axes2.plot(y, x)
```

```python
# Multi plot using subplots(nrows, ncols)
fig, new_axes = plt.subplots(nrows=1, ncols=2)
new_axes[0].set_title('plot 1')
new_axes[0].set_xlabel('x')
new_axes[0].set_ylabel('y')
new_axes[0].plot(x, y)

new_axes[1].plot(y, x)
plt.tight_layout()

# for a in new_axes:
#     a.plot(x, y)
```

```python
# save figure
fig.savefig('myplot.png')
```

```python
# Legend
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, x**2, label='x^2')
axes.plot(x, x**3, label='x^3')

axes.legend()
```

```python
# Show Grid 
axes.plot(x, x**2)
axes.grid(True)
fig
```

```python
# Colors and other customizatino
base = [0, 0, 1, 1]
fig = plt.figure()
ax = fig.add_axes(base)
ax.plot(x, y, color='purple', linestyle='--')

# axes limit
ax.set_xlim([0, 5])
ax.set_ylim([0, 10])
```
